Week 3 Task 2: Make a comparison to find the percentage of usage for customer and subscriber

```
# Loading the dataset
df = spark.read.table('week_2_output_csv')
```

```
df.show(20)
```

```
df.select('user_type').distinct().show() # if there were Null values this would list those also
```

```
import pyspark.sql.functions as F
```

```
df.groupBy('user_type').agg(F.round(F.count('user_type')/df.count()*100, 2).alias('User Percentage')).show()
```
